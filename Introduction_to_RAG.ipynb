### What is LLM? 

LLMs are essentially deep learning models which are basically collection of weights and biases. 

Formally, 
An LLM is a type of artificial intelligence trained on huge amounts of text so it can understand and generate human-like language.

Examples of LLMs include
- ChatGPT
- GPT - 4
- Gemini
- LLaMA


Most LLMs are built using a deep learning architecture called **Transformers**. 
Transformers usually follow Encoder - Decoder architecture.
