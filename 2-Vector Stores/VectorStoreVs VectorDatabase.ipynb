                 VECTOR STORE: 
- Vector store is a lightweight library or tool focussed on storing and searching vectors efficiently. 
- Core function : Simple similarity search - find the K nearest neighbours to a query vector
- Architecture : Usually runs in-memory or as a local file, single-machine operation. 
- Best for : Prototypes, research, small applications with <1M vectors. 
- Examples: FAISS, Annoy, ChromaDB, ScaNN, NMSUB 


                 VECTOR DATABASE: 
- Vector database is a full featured database system designed for vector data at scale. 
- Core function : Advanced search with filters, metadat queries, and database operations. 
- Architecture : Distributed system with replication, sharding and high availability.  
- Best for : Production systems, enterprise applications, billions of vectors. 
- Examples: Pinecone, Weaviate, Qdrant, Vespa, Milvus, Datastax

                     SIMPLE RULE OF CHOICE 

Start with a Vector Store for protoyping and leanring. 
Graduate to a Vector Database when you need prodcution-scale features, reliability, and advanced queryingcapabilities.
